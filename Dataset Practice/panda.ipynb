```python
import pandas as pd
# data frames
info = {
    "Name":["Raj","pradeep","Ronit"],
    "CGPA":["8.00","8.5","9.0"]
}
#we dont need to add index it auto labeled 
df = pd.DataFrame(info)
print(df)
```

```python
df2 = pd.DataFrame({
    "Data":[1],
    "Value":[10]
})
#always provide value in list
print(df2)
```

```python
#series 
s = pd.Series([1,2,3,4])
print(s)
print(s[0])
print(s[3])#index based access
```

```python
s = pd.Series([100,200,300,400],index=["a","b","c","d"])
print(s)
print(s["a"])
print(s["d"])#label bases access
```

```python
#data frame with list of list
df = pd.DataFrame([["Ronit",19],["jadu",21],["Hari om",36]],columns=["Name","age"],index=[100,200,300])
print(df)
print(df.index)
print(df.columns)
```
