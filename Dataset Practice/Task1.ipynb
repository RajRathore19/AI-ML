```python
import json 

with open("dataset.txt","r") as fs:
    data = json.load(fs)
   # print(data)
```

```python
def clean_data(data):
   for i in data:
       val = i.get("usage_hours")
       
       if (val == None):   
           i.update({"usage_hours":"20"})
      
       print(val)
        
clean_data(data)
```

```python
def check_status(data):
    for i in data:
        val = i.get("status").lower()
        if (val == "inactive"):
            i["status"]="active"
            
        print(val)

check_status(data)
```

```python
def check_id(data):
    for i in data:
        val = str(i.get("id"))
        val2 = set(val)
        print(val2)

check_id(data)
```
